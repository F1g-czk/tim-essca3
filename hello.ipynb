```java
import java.util.Scanner;

Scanner clavier = new Scanner(System.in);

int n;
do {
    System.out.println("Entrez un nombre entre 2 et 15 : ");
    n = clavier.nextInt();
    if (n < 2 || n > 15) {
        System.err.println("Erreur : la valeur doit être entre 2 et 15 !");
    }
} while (n < 2 || n > 15);

System.out.println("Carré creux (1) ou plein (2) ? ");
int choix = clavier.nextInt();

for (int ligne = 0; ligne < n; ligne++) {
    for (int colonne = 0; colonne < n; colonne++) {
        if (choix == 1) {
            // Carré creux
            if (ligne == 0 || ligne == n-1 || colonne == 0 || colonne == n-1) {
                System.out.print("* ");
            } else {
                System.out.print("  ");
            }
        } else {
            // Carré plein
            System.out.print("* ");
        }
    }
    System.out.println();
}
```

```java
import java.util.Scanner;

Scanner clavier = new Scanner(System.in);

int n;
do {
    System.out.println("Entrez un nombre entre 2 et 15 : ");
    n = clavier.nextInt();
    if (n < 2 || n > 15) {
        System.err.println("Erreur : la valeur doit être entre 2 et 15 !");
    }
} while (n < 2 || n > 15);

for (int ligne = 1; ligne <= n; ligne++) {
 
    for (int espace = 0; espace < n - ligne; espace++) {
        System.out.print(" ");
    }
    // Afficher les étoiles
    for (int etoile = 0; etoile < ligne; etoile++) {
        System.out.print("* ");
    }
    System.out.println();
}
```

```java
import java.util.Scanner;

Scanner clavier = new Scanner(System.in);

int n;
do {
    System.out.println("Entrez un nombre entre 2 et 15 : ");
    n = clavier.nextInt();
    if (n < 2 || n > 15) {
        System.err.println("Erreur : la valeur doit être entre 2 et 15 !");
    }
} while (n < 2 || n > 15);


for (int hauteurMax = 1; hauteurMax <= n; hauteurMax++) {
    for (int ligne = 1; ligne <= hauteurMax; ligne++) {
        // Afficher les espaces pour centrer
        for (int espace = 0; espace < n - ligne; espace++) {
            System.out.print(" ");
        }
   
        for (int etoile = 0; etoile < ligne; etoile++) {
            System.out.print("* ");
        }
        System.out.println();
    }
}
```
